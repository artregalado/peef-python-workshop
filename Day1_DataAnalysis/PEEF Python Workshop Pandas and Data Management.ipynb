```python
# Apply Function, also very important

#Change the Pclass to text
# First define a function to apply
def changeClass(column):
    if column == 1:
        pass_class = 'First class'
    elif column == 2:
        pass_class = 'Second class'
    else:
        pass_class = 'Third class'
    return pass_class

df['text_pclass'] = df['Pclass'].apply(changeClass)
df
```
