# Time Series Econometrics

Statsmodels is the main econometrics library for Python. 
As with the data analysis tool, the best place to know the capabilities and how to use the library is the documentation. 

+ [User guide](https://www.statsmodels.org/stable/user-guide.html)
+ [Time series econometrics](https://www.statsmodels.org/stable/tsa.html)

The second link will have information on how to estimate the common models you will find in the textbook and at the introductory-intermediate economics level. As has been the case throughout the workshop, the documentation is your best friend.

```python
# Time series relevant plots the ACF and the PACF
# We are gonna use some sample data
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

# Load and clean some sample data from statsmodels
dta = sm.datasets.sunspots.load_pandas().data
dta.index = pd.Index(sm.tsa.datetools.dates_from_range('1700', '2008'))
del dta["YEAR"]
dta
```

```python
dta.plot()
```

```python
# There is a bug from statsmodels when plotting to Jupyter Notebook that makes the plot be done twice
# Do not mind this bug. 
sm.graphics.tsa.plot_acf(dta.values.squeeze(), lags=40)
```

```python
sm.graphics.tsa.plot_pacf(dta.values.squeeze(), lags=40, method="ywm")
```

```python
# Check stationarity

y = dta.values.squeeze()

# Run the Dicky Fuller test
dftest = adfuller(y, autolag='AIC')

# Extract the test statistic
print('Test statistic: ', dftest[0])

# Extract the p-value
print('P-value: ', dftest[1])
```

```python
# We can use numpy.diff to compute first differences of an array and make it stationary
# Run the Dicky Fuller test
dftest = adfuller(np.diff(y), autolag='AIC')

# Extract the test statistic
print('Test statistic: ', dftest[0])

# Extract the p-value
print('P-value: ', dftest[1])
```

```python
# Check again the ACF and PACF
sm.graphics.tsa.plot_acf(np.diff(y), lags=40)
```

```python
sm.graphics.tsa.plot_pacf(np.diff(y), lags=40, method="ywm")
```

```python
# Building ARIMA models
y = dta.values.squeeze()
mod = sm.tsa.arima.ARIMA(y, order=(1, 1, 2))
res = mod.fit()
print(res.summary())
```

```python
fig, ax = plt.subplots()

ax.plot(y, label = "Original")
ax.plot(res.predict(), label = "Predicted")
ax.legend()
```
