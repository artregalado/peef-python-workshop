```python
# Load libraries
import glob                     #To get filenames and load data from various files
import re
import pandas as pd             #Data wrangling
import numpy as np              #For arrays and matrix
import matplotlib.pyplot as plt #Visualizations
import seaborn as sns           #Visualizations
import scipy.optimize as sco
%matplotlib inline              
```

```python
# Read files and transform into series
## Create a function that loads stock price data

def get_financial_series(file, date, adj_close, tsformat):
    '''
    Function to transform a CSV that has information of stock prices to
    a pandas series for future merging.
    File = filepath to the csv
    date = string with the name of the column where the date is stored
    adj_close = string with the name of the column where the adjusted price is
    format = string with the date format
    '''
    df = pd.read_csv(file)
    df[date] = pd.to_datetime(df[date], format= tsformat)
    df = df.set_index(date)
    series = df.loc[:,adj_close]
    return series
```

```python
## Because there can be many series in CSV files
## Apply a for loop to use the function on each file

financialseries = glob.glob('./datos_financieros/*.csv')  # Get all csv files containing financial series
tsformat = '%Y/%m/%d'                                     # Date format on the series
series = dict()                                           # Initialize an empty dict to store series

for file in financialseries:
    # Get only the stock names from the files
    string = file
    string = string.split('\\')[1]
    lookup = re.search('.csv', string)
    string = string[0:lookup.start()]
    # Apply the function to each stock file
    x = get_financial_series(file,'Date','Adj Close', tsformat)
    #Save the Adj Close series into a dict where the key value is
    #the stock name retrieved in the first part of the code
    series[string] = x

series = pd.DataFrame(series)
print('--------------Dataframe information---------------')
print(series.info())

print('--------------First observartions---------------')
print(series.head())
    
```

```python
# Plot the series in absolute values
sns.set_style('darkgrid')
plt.figure(figsize=(14, 7))
for c in series.columns.values:
    plt.plot(series.index, series[c], lw=3, alpha=0.8,label=c)
plt.legend(loc='upper left', fontsize=12)
plt.ylabel('price in MXN')
```

```python
# Make a function to get anualized returns and simulate different portfolio weights to find most efficient portfolio
def portfolio_annualised_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(mean_returns*weights ) *252
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
    return std, returns


def random_portfolios(num_portfolios, mean_returns, cov_matrix, risk_free_rate):

    results = np.zeros((3,num_portfolios))
    weights_record = []
    for i in np.arange(num_portfolios):
        weights = np.random.random(3)
        weights /= np.sum(weights)
        weights_record.append(weights)
        portfolio_std_dev, portfolio_return = portfolio_annualised_performance(weights, mean_returns, cov_matrix)
        results[0,i] = portfolio_std_dev
        results[1,i] = portfolio_return
        results[2,i] = (portfolio_return - risk_free_rate) / portfolio_std_dev
    return results, weights_record
```

```python
def display_simulated_ef_with_random(mean_returns, cov_matrix, num_portfolios, risk_free_rate):

    results, weights = random_portfolios(num_portfolios,mean_returns, cov_matrix, risk_free_rate)

    

    max_sharpe_idx = np.argmax(results[2])

    sdp, rp = results[0,max_sharpe_idx], results[1,max_sharpe_idx]

    max_sharpe_allocation = pd.DataFrame(weights[max_sharpe_idx],columns=['allocation'])

    max_sharpe_allocation.allocation = [round(i*100,2)for i in max_sharpe_allocation.allocation]

    max_sharpe_allocation = max_sharpe_allocation.T

    

    min_vol_idx = np.argmin(results[0])

    sdp_min, rp_min = results[0,min_vol_idx], results[1,min_vol_idx]

    min_vol_allocation = pd.DataFrame(weights[min_vol_idx],columns=['allocation'])

    min_vol_allocation.allocation = [round(i*100,2)for i in min_vol_allocation.allocation]

    min_vol_allocation = min_vol_allocation.T

    print ("-"*80)

    print ("Maximum Sharpe Ratio Portfolio Allocation\n")

    print ("Annualised Return:", round(rp,2))

    print ("Annualised Volatility:", round(sdp,2))

    print ("\n")

    print (max_sharpe_allocation)

    print ("-"*80)

    print ("Minimum Volatility Portfolio Allocation\n")

    print ("Annualised Return:", round(rp_min,2))

    print ("Annualised Volatility:", round(sdp_min,2))

    print ("\n")

    print (min_vol_allocation)

    

    plt.figure(figsize=(10, 7))

    plt.scatter(results[0,:],results[1,:],c=results[2,:],cmap='YlGnBu', marker='o', s=10, alpha=0.3)

    plt.colorbar()
    plt.scatter(sdp,rp,marker='*',color='r',s=500, label='Maximum Sharpe ratio')

    plt.scatter(sdp_min,rp_min,marker='*',color='g',s=500, label='Minimum volatility')

    plt.title('Simulated Portfolio Optimization based on Efficient Frontier')

    plt.xlabel('annualised volatility')

    plt.ylabel('annualised returns')

    plt.legend(labelspacing=0.8)
```

```python
#Simulate different portfolio weights and search for efficient
mean_returns = series_returns.mean()
cov_matrix = series_returns.cov()
num_portfolios = 25000
risk_free_rate = 0.08
display_simulated_ef_with_random(mean_returns, cov_matrix, num_portfolios, risk_free_rate)
```
